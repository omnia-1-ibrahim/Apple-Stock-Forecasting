```python
# استيراد المكتبات
import yfinance as yf
import pandas as pd

# تحميل بيانات سهم Apple من 1 يناير 2018 حتى 31 ديسمبر 2023
df = yf.download("AAPL", start="2018-01-01", end="2024-01-01")

# عرض أول 5 صفوف
df.head()
```

```python
print(df)
```

```python
# 2. التحليل الأولي
print(df.describe())
df['Close'].plot(title="Apple Stock Price (5 Years)")
```

```python
# 3. التحليل الموسمي
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(df['Close'], model='additive', period=365)
decomposition.plot()
```

```python
# إعادة تعيين الـ index لو التاريخ موجود فيه
df.reset_index(inplace=True)

# نحتفظ فقط بالتاريخ وسعر الإغلاق
df = df[['Date', 'Close']]

# إعادة تسمية الأعمدة بالطريقة اللي Prophet بيشتغل بيها
df.rename(columns={'Date': 'ds', 'Close': 'y'}, inplace=True)

# عرض أول 5 صفوف للتأكد
df.head()
```

```python
print(df.dtypes)
print(df.isnull().sum())
df.head()
```

```python
# إزالة الطبقة العلوية من MultiIndex (Ticker AAPL)
df.columns = df.columns.get_level_values(0)

# دلوقتي إعادة التسمية بطريقة Prophet
df.rename(columns={'Date': 'ds', 'Close': 'y'}, inplace=True)

# تأكيد النوع الصحيح
df['ds'] = pd.to_datetime(df['ds'])
df['y'] = pd.to_numeric(df['y'], errors='coerce')
df.dropna(inplace=True)

# مراجعة النتيجة
df.head()
```

```python
from prophet import Prophet

model = Prophet()
model.fit(df)
```

```python
# إنشاء DataFrame فيه تواريخ المستقبل (مثلاً 30 يوم)
future = model.make_future_dataframe(periods=30)

# توليد التوقعات
forecast = model.predict(future)

# عرض أول صفوف من التوقعات
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
model.plot(forecast);
```

```python
model.plot_components(forecast);
```

```python
# دمج القيم المتوقعة مع القيم الحقيقية
df_forecast = forecast[['ds', 'yhat']].merge(df, on='ds', how='left')

# حذف أي قيم ناقصة (لو موجودة)
df_forecast.dropna(inplace=True)

df_forecast.head()
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# حساب RMSE بين القيمة الحقيقية والمتوقعة
rmse = np.sqrt(mean_squared_error(df_forecast['y'], df_forecast['yhat']))
print(f"RMSE: {rmse:.2f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(df_forecast['ds'], df_forecast['y'], label='Actual Price', color='black')
plt.plot(df_forecast['ds'], df_forecast['yhat'], label='Predicted Price', color='blue')
plt.legend()
plt.title('Actual vs. Predicted Apple Stock Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# تحميل البيانات
df_lstm = df.copy()

# نشتغل بس على عمود السعر
data = df_lstm[['y']].values

# نعمل Scaling بين 0 و 1
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

# نحدد window size
sequence_length = 60

# نجهز الداتا كـ X و y
X = []
y = []
for i in range(sequence_length, len(scaled_data)):
    X.append(scaled_data[i-sequence_length:i, 0])
    y.append(scaled_data[i, 0])

X = np.array(X)
y = np.array(y)

# نعيد تشكيل الـ X علشان LSTM يتعامل معها (samples, time_steps, features)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout

model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(units=1))  # سعر واحد في كل مرة

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(X, y, epochs=10, batch_size=32)
```

```python
predicted_prices = model.predict(X)
predicted_prices = scaler.inverse_transform(predicted_prices)  # نرجّعها للقيم الأصلية

real_prices = scaler.inverse_transform(y.reshape(-1, 1))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(real_prices, color='black', label='Actual Price')
plt.plot(predicted_prices, color='green', label='LSTM Predicted Price')
plt.title('LSTM Prediction vs Actual Apple Price')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
rmse_lstm = np.sqrt(mean_squared_error(real_prices, predicted_prices))
print(f"LSTM RMSE: {rmse_lstm:.2f}")
```

```python
# لازم تكوني نفذتي Prophet (df_forecast) و LSTM (predicted, actual)

# تجهيز البيانات للمقارنة
comparison_df = pd.DataFrame({
    'Date': df['ds'][-len(y):].reset_index(drop=True),
    'Actual': scaler.inverse_transform(y.reshape(-1, 1)).flatten(),
    'Prophet': df_forecast['yhat'][-len(y):].values,
    'LSTM': scaler.inverse_transform(predicted_prices).flatten()
})

# رسم المقارنة
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))
plt.plot(comparison_df['Date'], comparison_df['Actual'], label='Actual', color='black')
plt.plot(comparison_df['Date'], comparison_df['Prophet'], label='Prophet', color='blue')
plt.plot(comparison_df['Date'], comparison_df['LSTM'], label='LSTM', color='green')
plt.title("Comparison: Actual vs Prophet vs LSTM")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# حفظ الملف بصيغة CSV
comparison_df.to_csv("forecast_comparison.csv", index=False)
```
